```python
from spore.model.spatial_dist import PureClustering_Spherical, PoissonClustering_Spherical
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

from healpy.visufunc import mollview

import numpy as np
```

```python
sphsky = PureClustering_Spherical(f0=1., sky_size=1.0, ncells=256, power_spectrum = lambda l : (l/0.01)**-2.)
```

```python
mollview(sphsky.healpix_brightness([10.])[0])
```

```python
plt.imshow(sphsky.sky([10.])[0])
```

```python
psphsky = PoissonClustering_Spherical(f0=1., sky_size=1.0, ncells=256, power_spectrum = lambda l : (l)**-3.)
```

```python
mollview(1+psphsky._healpix_deltax)
```

```python
source_pos_ang = psphsky.source_positions_ang(1000.)
```
